Scale factor corresponds to: 
two factors of the photon SF for differences in the ID efficiency between data and MC, 
two factors of the photon SF specifically for the efficiency of the pixel seed veto requirement, 
one factor of the triggers efficiency wrt invariant mass, 
one factor of the efficiency of the leading leg of the trigger, and 
two factors of the efficiency of the triggers trailing leg.

````python
masses = [1400,1450,1500,1550,1600,1650,1700,1750,1800,1850,1900,1950,2000,2050,2100,
          2200,2300,2400,2500]
neutralino_masses = [10,25,50,100,150,200,300,400,500,600,700,800,900,1000,1100,1200,
                     1300,1400,1500,1600,1700,1800,1900,2000,2100,2200,2300,2400]

scale_factor = 1.002*1.002*0.998*0.998*0.994*0.982*0.998*0.998


h1.Scale(scale_factor)
h2.Scale(scale_factor)
h3.Scale(scale_factor)
h4.Scale(scale_factor)
h5.Scale(scale_factor)
h6.Scale(scale_factor)
````

````python
import re
import os
import math

if cf_name == 'T5gg':
    temp_file = open('./countingCrossCheckCentral_XXX_YYY.txt', 'r')
    cf_re = 'T5GG'
elif cf_name == 'T6gg':
    temp_file = open('./counting_T6Wg_XXX_YYY.txt', 'r')
    cf_re = 'T6GG'
    
lines = [line for line in temp_file.readlines()]

cfpath = './countingfiles_'+ cf_re # + '_ext'

# Make the ./counting_files dir
# If already there rm the files
try:
    os.mkdir(cfpath)
except OSError:
    cfiles = os.listdir(cfpath)
    for cf in cfiles:
        os.remove(cfpath+'/'+cf)
        print 'removing ', cfpath+'/'+cf
````

````python
# See above note
f = 1.0

for it_i in range(1,len(masses)+1):
    temp_masses = neutralino_masses
    if masses[it_i-1]%100:
        temp_masses = temp_masses+[masses[it_i-1] - 300,masses[it_i-1]-200,
                                   masses[it_i-1]-100,masses[it_i-1]-25,masses[it_i-1]-10]
    else:
        temp_masses = temp_masses+[masses[it_i-1] - 250,masses[it_i-1]-150,
                                   masses[it_i-1]-50,masses[it_i-1]-25,masses[it_i-1]-10]
    for it_j in range(1,len(temp_masses)+1):
        i = int(h1.GetXaxis().FindBin(masses[it_i-1]) )
        j = int(h1.GetYaxis().FindBin(temp_masses[it_j -1]) )
        
        #This should cover most cases where statistics are limited
        if h1.GetBinContent(i,j) > 0.0 and temp_masses[it_j -1] < masses[it_i-1]:
            #hack to deal with bins that have events in the first signal bin but not later bins 
            if temp_masses[it_j -1] == 1950 and masses[it_i-1] == 2050 and t5:
                continue;
            if temp_masses[it_j -1] == 1700 and masses[it_i-1] == 2300 and t5:
                continue;
            if temp_masses[it_j -1] == 1700 and masses[it_i-1] == 2400 and t5:
                continue;
            if temp_masses[it_j -1] == 2200 and masses[it_i-1] == 2500 and t5:
                continue;
         
            uncertUp1 = math.fabs(h1up.GetBinContent(i,j))
            uncertUp2 = math.fabs(h2up.GetBinContent(i,j))
            uncertUp3 = math.fabs(h3up.GetBinContent(i,j))
            uncertUp4 = math.fabs(h4up.GetBinContent(i,j))
            uncertUp5 = math.fabs(h5up.GetBinContent(i,j))
            uncertUp6 = math.fabs(h6up.GetBinContent(i,j))
            
            uncertDown1 = math.fabs(h1down.GetBinContent(i,j))
            uncertDown2 = math.fabs(h2down.GetBinContent(i,j))
            uncertDown3 = math.fabs(h3down.GetBinContent(i,j))
            uncertDown4 = math.fabs(h4down.GetBinContent(i,j))
            uncertDown5 = math.fabs(h5down.GetBinContent(i,j))
            uncertDown6 = math.fabs(h6down.GetBinContent(i,j))
            
            file_name = 'counting_'+ cf_re +'_'+ str(masses[it_i-1]) +'_'+ str(temp_masses[it_j-1]) +'.txt'
            file_name = cfpath+'/'+file_name
            new_file = open(file_name, 'w')
            print 'writing ', file_name
            
            for line in lines:
                
                if re.search(cf_re, line):
                    l = line.replace(cf_re+'1', str(f*round(h1.GetBinContent(i,j),3)))
                    l = l.replace(cf_re+'2', str(f*round(h2.GetBinContent(i,j),3))) 
                    l = l.replace(cf_re+'3', str(f*round(h3.GetBinContent(i,j),3)))
                    l = l.replace(cf_re+'4', str(f*round(h4.GetBinContent(i,j),3)))
                    l = l.replace(cf_re+'5', str(f*round(h5.GetBinContent(i,j),3)))
                    l = l.replace(cf_re+'6', str(f*round(h6.GetBinContent(i,j),3)))
                    new_file.write(l)
                    
                elif re.search('MC', line):
                    l = line.replace('MCN1', str(int(round(h1uw.GetBinContent(i,j),4))))
                    l = l.replace('MCN2', str(int(round(h2uw.GetBinContent(i,j),4))))
                    l = l.replace('MCN3', str(int(round(h3uw.GetBinContent(i,j),4))))
                    l = l.replace('MCN4', str(int(round(h4uw.GetBinContent(i,j),4))))
                    l = l.replace('MCN5', str(int(round(h5uw.GetBinContent(i,j),4))))
                    l = l.replace('MCN6', str(int(round(h6uw.GetBinContent(i,j),4))))
                    
                    l = l.replace('MCA1', str(round((f*h1.GetBinContent(i,j)) /  h1uw.GetBinContent(i,j),4)))
                    l = l.replace('MCA2', str(round((f*h2.GetBinContent(i,j)) /  h2uw.GetBinContent(i,j),4)))
                    l = l.replace('MCA3', str(round((f*h3.GetBinContent(i,j)) /  h3uw.GetBinContent(i,j),4)))
                    l = l.replace('MCA4', str(round((f*h4.GetBinContent(i,j)) /  h4uw.GetBinContent(i,j),4)))
                    l = l.replace('MCA5', str(round((f*h5.GetBinContent(i,j)) /  h5uw.GetBinContent(i,j),4)))
                    l = l.replace('MCA6', str(round((f*h6.GetBinContent(i,j)) /  h6uw.GetBinContent(i,j),4)))
                    
                    #print f*h1.GetBinContent(i,j), h1uw.GetBinContent(i,j), f*h1.GetBinContent(i,j)/h1uw.GetBinContent(i,j)
                    #print f*h2.GetBinContent(i,j), h2uw.GetBinContent(i,j), f*h2.GetBinContent(i,j)/h2uw.GetBinContent(i,j)
                    
                    new_file.write(l)
                
                elif re.search('JES', line):
                    l = line.replace('JES1',  str(round(1.0 + max(uncertUp1, uncertDown1),4))) 
                    l = l.replace('JES2',  str(round(1.0 + max(uncertUp2, uncertDown2),4)))
                    l = l.replace('JES3',  str(round(1.0 + max(uncertUp3, uncertDown3),4)))
                    l = l.replace('JES4',  str(round(1.0 + max(uncertUp4, uncertDown4),4)))
                    l = l.replace('JES5',  str(round(1.0 + max(uncertUp5, uncertDown5),4)))
                    l = l.replace('JES6',  str(round(1.0 + max(uncertUp6, uncertDown6),4)))
                    new_file.write(l)
                
                else:
                    new_file.write(line)
                    
            new_file.close()
````

Now the datacards should be in the counting_files dir and make the shell script to run the combine commands which follow the template:
```
combine -M Asymptotic counting_files/counting_t6gg_${i}_${j}.txt > limit_files/limit_t6gg_${i}_${j}.txt
```

````python
blind = True

sh_file_path = 'makelimitfiles_' + cf_re

if blind:
    sh_file_path += '_blind'
else: 
    sh_file_path += '_observed'

sh_file_path = sh_file_path +'_ext.sh'
sh_file = open(sh_file_path, 'w')

for cf in os.listdir(cfpath):
        
    m1 = cf.split('_')[2]
    m2 = cf.split('_')[3][:-4]
        
    print m1, m2
    
    command = 'combine -M Asymptotic '
    
    if blind:
        command += '--run blind '
    else:
        command += '--run observed '
    command += cfpath +'/' + cf + ' > limitfiles_' + cf_re 
    
   
    if blind:
        command += '_blind'
    else: 
        command += '_observed'
    command += '/'+'limit_'+cf_re+'_'+m1+'_'+m2+'.txt'
        
    print command
        
    sh_file.write('echo "'+ command +'"\n')
    sh_file.write(command+'\n')

sh_file.close()
````
